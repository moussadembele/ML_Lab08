```python
import math

# Function to calculate Euclidean distance between two 3D points
def calculate_distance(x1, y1, z1, x2, y2, z2):
    distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2 + (z2 - z1)**2)
    return distance

# Taking input from the user
print("Enter the coordinates of the first point (x1, y1, z1):")
x1 = float(input("x1: "))
y1 = float(input("y1: "))
z1 = float(input("z1: "))

print("Enter the coordinates of the second point (x2, y2, z2):")
x2 = float(input("x2: "))
y2 = float(input("y2: "))
z2 = float(input("z2: "))

distance = calculate_distance(x1, y1, z1, x2, y2, z2)

print(f"The distance between the two points is: {distance:.2f}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
iris_data = iris.data
iris_target = iris.target
iris_feature_names = iris.feature_names
iris_target_names = iris.target_names

df = pd.DataFrame(iris_data, columns=iris_feature_names)
df['target'] = iris_target

df_filtered = df[df['target'] != 2]  # Keep only class 0 and class 1

attribute1 = 'sepal length (cm)'
attribute2 = 'sepal width (cm)'

plt.figure(figsize=(8, 6))

plt.scatter(df_filtered[df_filtered['target'] == 0][attribute1],
            df_filtered[df_filtered['target'] == 0][attribute2],
            color='blue', label=iris_target_names[0])

plt.scatter(df_filtered[df_filtered['target'] == 1][attribute1],
            df_filtered[df_filtered['target'] == 1][attribute2],
            color='red', label=iris_target_names[1])

plt.xlabel(attribute1)
plt.ylabel(attribute2)

plt.legend()

plt.title(f'Scatter plot of {attribute1} vs {attribute2}')
plt.show()
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Sat Sep 14 10:08:26 2024

@author: Vijayakrishna
"""

# Let's first import the Iris dataset and then calculate the Euclidean Distance Matrix for all the data samples in the feature space.

from sklearn.datasets import load_iris
from scipy.spatial.distance import pdist, squareform
import pandas as pd

# Load Iris dataset
iris = load_iris()

# Print attributes and methods of an object
print(dir(iris))

# Feature names (attributes)
print("Feature Names:", iris.feature_names)

# Target names (species)
print("Target Names:", iris.target_names)

# Dataset description
print("Description:\n", iris.DESCR)

# First 5 rows of data (features)
print("\nFirst 5 data samples:\n", iris.data[:5])

# First 5 target labels (species)
print("\nFirst 5 target labels:\n", iris.target[:5])


data = iris.data  # we only need the features for distance calculation

# Compute pairwise Euclidean distance
distance_matrix = squareform(pdist(data, metric='euclidean'))

# Create a DataFrame for better readability
distance_df = pd.DataFrame(distance_matrix)

# Display the distance matrix
print(distance_df.head(10))
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Sat Sep 14 10:32:52 2024

@author: Vijayakrishna
"""

# Let's import the Iris dataset and filter it to include samples from only two classes. 
# We will then create a scatter plot of two chosen input attributes (e.g., sepal length and petal length).

import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
import pandas as pd

# Load Iris dataset
iris = load_iris()
data = iris.data
target = iris.target

# Create a DataFrame for easier manipulation
df = pd.DataFrame(data, columns=iris.feature_names)
df['target'] = target

# Select only two classes (0: Setosa, 1: Versicolor) and two attributes (sepal length and petal length)
filtered_df = df[df['target'].isin([2, 1])] #([0, 1])]

# Select the attributes for scatter plot (sepal length and petal length)
x = filtered_df['sepal length (cm)']
y = filtered_df['petal length (cm)']
labels = filtered_df['target']

# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, c=labels, cmap='viridis', edgecolor='k', s=100)
plt.title('Scatter Plot of Two Iris Classes (Setosa vs Versicolor)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.grid(True)

# Manual Boundary line to classify the two groups
plt.plot([4.5, 8],[4.5,5.5], 'ro-')

# Show the plot
plt.show()

# List all the colormaps inside matplotlib
import matplotlib.cm as cm

color_maps = list(cm.cmaps_listed.keys())
print(color_maps)
```
